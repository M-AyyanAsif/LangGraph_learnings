```python
from langgraph.graph import StateGraph, START, END
from langchain_google_genai import ChatGoogleGenerativeAI
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
model = ChatGoogleGenerativeAI(
    api_key = "AIzaSyCjqWFAlt7es-_0TCSupT-BpZMYYnl5Wn0",
    model = "gemini-3-flash-preview",
    temperature = 0.7
)
```

```python
class BlogState(TypedDict):
    title: str
    outline: str
    content: str
```

```python
def create_outline(state: BlogState) ->BlogState:
    title = state['title']

    prompt  = f"Generate a detalied outline for the blog on the following topic -{title}"
    outline = model.invoke(prompt).content

    state['outline'] = outline
    
    return state
```

```python
def create_blog(state: BlogState) -> BlogState:

    title = state['title']
    outline = state['outline']

    prompt = f'Write a detailed blog on the title - {title} using the follwing outline \n {outline}'

    content = model.invoke(prompt).content

    state['content'] = content

    return state
```

```python
Graph = StateGraph(BlogState)

Graph.add_node('create_outline', create_outline)
Graph.add_node('CreateBlog', create_blog)

Graph.add_edge(START, 'create_outline')
Graph.add_edge('create_outline', 'CreateBlog')
Graph.add_edge('CreateBlog', END)

response = Graph.compile()
```

```python
initial_state = {'title': 'Rise of AI in Pkaistan'}
final_state = response.invoke(initial_state)

print(final_state)
```
